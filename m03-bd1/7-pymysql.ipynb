```python
# Criando um cursor

cursor = conexao.cursor()
```

Vamos criar duas tabelas:
*   categoria: id_categoria, nome_categoria
*   produto: id_produto, nome_produto

```python
ddl_categoria = """ create table categoria (
                      id_categoria int auto_increment primary key,
                      nome_categoria varchar(200)
                    );
                """

cursor.execute(ddl_categoria)
```

```python
ddl_produto = """ create table produto (
                    id_produto int auto_increment primary key,
                    id_categoria int,
                    nome_produto varchar(200) not null,
                    foreign key (id_categoria) references categoria (id_categoria)
                  );
              """

cursor.execute(ddl_produto)
```

```python
dml_insert_categoria = """ insert into categoria (nome_categoria) values
                              ("limpeza"),
                              ("fruta");
                       """

cursor.execute(dml_insert_categoria)
```

```python
# Para atribuir os valores adicionados, precisamos realizar um commit:

conexao.commit()
```
