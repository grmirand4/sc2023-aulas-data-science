## Conceitos Básicos de Probabilidade

### Fundamentos Teóricos

A probabilidade é uma área da matemática e estatística que estuda a chance de ocorrência de eventos. Ela é utilizada para analisar situações em que os resultados são incertos ou aleatórios.

Alguns conceitos fundamentais da probabilidade são:

1. Espaço Amostral: É o conjunto de todos os resultados possíveis de um experimento aleatório. Por exemplo, no lançamento de um dado, o espaço amostral é {1, 2, 3, 4, 5, 6}.

2. Evento: É um subconjunto do espaço amostral, ou seja, é um conjunto de resultados possíveis. Por exemplo, no lançamento de um dado, o evento "sair um número par" é o conjunto {2, 4, 6}.

3. Probabilidade: É uma medida numérica que indica a chance de ocorrência de um evento. A probabilidade de um evento A, denotada por P(A), é um número entre 0 e 1, onde 0 significa impossibilidade e 1 significa certeza. Por exemplo, a probabilidade de sair um número par no lançamento de um dado é 3/6 ou 1/2.

4. Regra da Soma: Sejam A e B dois eventos mutuamente exclusivos, ou seja, que não podem ocorrer ao mesmo tempo. A probabilidade da união desses eventos, denotada por P(A ∪ B), é igual à soma das probabilidades de cada evento individualmente. Por exemplo, se A é o evento "sair um número par" e B é o evento "sair um número ímpar" no lançamento de um dado, então P(A ∪ B) = P(A) + P(B) = 1/2 + 1/2 = 1.

5. Regra do Produto: Sejam A e B dois eventos independentes, ou seja, a ocorrência de um evento não afeta a ocorrência do outro. A probabilidade da interseção desses eventos, denotada por P(A ∩ B), é igual ao produto das probabilidades de cada evento individualmente. Por exemplo, se A é o evento "sair um número par" e B é o evento "sair um número maior que 3" no lançamento de um dado, então P(A ∩ B) = P(A) * P(B) = 1/2 * 1/2 = 1/4.

## Propriedades de Eventos de Probabilidade

### Fundamentos Teóricos

A probabilidade é uma medida numérica que descreve a chance de um evento ocorrer. Para que uma função P seja considerada uma probabilidade, ela deve satisfazer algumas propriedades:

1. A probabilidade de um evento A ocorrer é um número entre 0 e 1, inclusive. Ou seja, 0 ≤ P(A) ≤ 1.

2. A probabilidade do espaço amostral Ω (conjunto de todos os eventos possíveis) é igual a 1. Ou seja, P(Ω) = 1.

3. Se A e B são eventos mutuamente exclusivos (ou seja, não podem ocorrer ao mesmo tempo), então a probabilidade da união desses eventos é igual à soma das probabilidades individuais. Ou seja, P(A ∪ B) = P(A) + P(B).

### Exemplo em Python

Suponha que queremos calcular a probabilidade de obter um número par ao lançar um dado de seis faces. Podemos usar a seguinte função em Python para calcular essa probabilidade:

```python
def probabilidade_numero_par():
    total_possibilidades = 6
    numero_pares = 3
    probabilidade = numero_pares / total_possibilidades
    return probabilidade

resultado = probabilidade_numero_par()
print(f"A probabilidade de obter um número par é {resultado}")
```

Resposta: A probabilidade de obter um número par é 0.5.

## Métodos de Avaliação da Probabilidade

### Fundamentos Teóricos

A probabilidade é uma medida numérica que representa a chance de um evento ocorrer. Existem duas maneiras de calcular a probabilidade de um evento:

1. A partir das propriedades teóricas do evento: Nesse caso, é possível calcular a probabilidade de um evento com base em suas características físicas ou propriedades conhecidas. Por exemplo, ao lançar um dado perfeito e homogêneo, onde todas as faces são simétricas e têm igual chance de ocorrer, a probabilidade de cada face do dado é igual a 1/6 ou aproximadamente 0,166.

2. A partir de experimentações exaustivas: Em alguns casos, é necessário realizar experimentos repetidos para determinar a probabilidade de um evento. Por exemplo, ao lançar um dado várias vezes e anotar os resultados, é possível calcular a frequência relativa de cada face do dado e utilizar esses valores como uma estimativa da probabilidade. Quanto mais experimentos forem realizados, mais precisos serão os resultados.

### Exemplo em Python

A seguir, temos um exemplo em Python que simula 1 milhão de lançamentos de um dado e calcula as frequências relativas de cada face:

```python
import numpy as np
import pandas as pd

# Definindo uma semente aleatória
np.random.seed(42)

# Número de experimentos
N = 1000000

# Definindo os resultados de um dado a partir de números aleatórios entre 1 e 6
amostra = np.random.randint(low=1, high=7, size=N)

# Convertendo a amostra para uma Série e calculando as frequências relativas
frequencias = pd.Series(amostra).value_counts(normalize=True)
print(frequencias)
```

## Probabilidade da União de Eventos

### Fundamentos Teóricos

A probabilidade da união de dois eventos A e B é calculada pela fórmula:

P(A ∪ B) = P(A) + P(B) - P(A ∩ B)

Isso significa que a probabilidade da união de A e B é igual à soma das probabilidades de A e B, subtraindo a probabilidade da interseção entre eles.

### Exemplo em Python

Suponha que estamos analisando o desempenho de estudantes em duas disciplinas: Matemática e Física. Queremos calcular a probabilidade de um estudante ser aprovado em pelo menos uma dessas disciplinas.

```python
# Dados
total_alunos = 100
aprovados_matematica = 70
aprovados_fisica = 60
aprovados_ambas = 40

# Cálculo das probabilidades
prob_aprovados_matematica = aprovados_matematica / total_alunos
prob_aprovados_fisica = aprovados_fisica / total_alunos
prob_aprovados_ambas = aprovados_ambas / total_alunos

# Cálculo da probabilidade da união
prob_uniao = prob_aprovados_matematica + prob_aprovados_fisica - prob_aprovados_ambas

print("A probabilidade de um estudante ser aprovado em pelo menos uma das disciplinas é:", prob_uniao)
```

## Eventos Independentes

### Fundamentos Teóricos

Na teoria da probabilidade, dois eventos A e B são considerados independentes se a ocorrência de um deles não afeta a probabilidade de ocorrência do outro. Em outras palavras, a ocorrência de um evento não influencia a ocorrência do outro.

A probabilidade condicional de um evento A dado um evento B é calculada pela fórmula:

P(A | B) = P(A ∩ B) / P(B)

Se A e B são eventos independentes, então a probabilidade conjunta de A e B é calculada pela fórmula:

P(A ∩ B) = P(A) * P(B)

### Exemplo em Python

Suponha que você esteja jogando um dado honesto de 6 faces. Vamos considerar os eventos A e B:

A: O número sorteado é par
B: O número sorteado é maior que 3

Vamos calcular a probabilidade de A e B ocorrerem simultaneamente, considerando que A e B são eventos independentes.

```python
# Importando a biblioteca random
import random

# Definindo a função para simular o lançamento do dado
def lancar_dado():
    return random.randint(1, 6)

# Definindo a função para verificar se o número é par
def eh_par(numero):
    return numero % 2 == 0

# Definindo a função para verificar se o número é maior que 3
def eh_maior_que_3(numero):
    return numero > 3

# Simulando o lançamento do dado
numero_sorteado = lancar_dado()

# Verificando se o número é par
if eh_par(numero_sorteado):
    evento_A = True
else:
    evento_A = False

# Verificando se o número é maior que 3
if eh_maior_que_3(numero_sorteado):
    evento_B = True
else:
    evento_B = False

# Verificando se os eventos A e B ocorreram simultaneamente
if evento_A and evento_B:
    evento_A_e_B = True
else:
    evento_A_e_B = False

# Calculando a probabilidade de A e B ocorrerem simultaneamente
probabilidade_A_e_B = 1/2 * 1/2

# Imprimindo os resultados
print("Número sorteado:", numero_sorteado)
print("Evento A:", evento_A)
print("Evento B:", evento_B)
print("Evento A e B:", evento_A_e_B)
print("Probabilidade de A e B ocorrerem simultaneamente:", probabilidade_A_e_B)
```

Resposta:
```
Número sorteado: 4
Evento A: True
Evento B: True
Evento A e B: True
Probabilidade de A e B ocorrerem simultaneamente: 0.25
```

## Teorema de Bayes

### Fundamentos Teóricos

O Teorema de Bayes é uma ferramenta estatística que nos permite atualizar a probabilidade de um evento ocorrer, com base em novas informações. Ele é muito útil em situações onde temos uma hipótese inicial e queremos ajustá-la com base em evidências adicionais.

O teorema é definido da seguinte forma:

P(A|B) = (P(B|A) * P(A)) / P(B)

Onde:
- P(A|B) é a probabilidade do evento A ocorrer dado que o evento B ocorreu.
- P(B|A) é a probabilidade do evento B ocorrer dado que o evento A ocorreu.
- P(A) é a probabilidade do evento A ocorrer.
- P(B) é a probabilidade do evento B ocorrer.

O Teorema de Bayes é muito utilizado em áreas como medicina, inteligência artificial, finanças e muitas outras, onde é necessário atualizar probabilidades com base em novas informações.

### Exemplo em Python com resposta

Suponha que temos uma caixa com 100 bolas, sendo 70 vermelhas e 30 azuis. Vamos calcular a probabilidade de uma bola ser vermelha, dado que ela foi escolhida aleatoriamente da caixa.

```python
# Importando a biblioteca numpy
import numpy as np

# Definindo as probabilidades
P_vermelha = 70/100
P_azul = 30/100

# Calculando a probabilidade condicional
P_vermelha_dado_escolha = (P_vermelha * 1) / (P_vermelha * 1 + P_azul * 0)

# Imprimindo o resultado
print("A probabilidade de uma bola ser vermelha, dado que ela foi escolhida aleatoriamente da caixa, é de:", P_vermelha_dado_escolha)
```

Output:
```
A probabilidade de uma bola ser vermelha, dado que ela foi escolhida aleatoriamente da caixa, é de: 0.7
```
