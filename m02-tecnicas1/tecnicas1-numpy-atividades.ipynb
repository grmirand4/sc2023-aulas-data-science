### Atividade 3

Considere um sistema simples de notas de uma escola. Neste sistema, para cada aluno, há um registro de nome e uma nota.
Crie uma lista que contenha os nomes de pelo menos 5 alunos e um array numpy com suas respectivas notas. O sistema de exibir o nome e nota dos alunos com nota maior ou igual a 7.

```python
import numpy as np

nomes = np.array(["Joao", "Maria", "Fernando", "Francisco", "Gabriel"])
notas = np.array([9,6,1,7,10])

filtro = notas >= 7

for nome, nota in zip(nomes[filtro], notas[filtro]):
    print(f"{nome}: {nota}")
```

Faça um programa que crie dois arrays numpy com valores numéricos. O programa deve verificar e criar um terceiro array com os valores em comum entre os dois anteriores.

```python
%%time
array1 = np.array([1,2,23,2,4,5])
array2 = np.array([10,2,2,7,4,6])

filtro = np.array([], dtype = bool)
filtro = np.append(filtro, [n in array2 for n in array1])

array3 = np.unique(array1[filtro])

print(array3)
```

```python
%%time
# Forma mais simples:

np.intersect1d(array1, array2)
```
