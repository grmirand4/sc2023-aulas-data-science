### Merge

##### Modos de mesclar

![](https://s3-sa-east-1.amazonaws.com/lcpi/910ed9dc-2768-47c3-9f21-ade9e156eaa7.png)

**1. inner**: mescla linhas cujas chaves estejam em ambos os dataframes.<br>
**2. outer**: mescla todas as linhas de ambos os dataframes.<br>
**3. left**: mescla todas as linhas de y que estiverem contidas em x.<br>
**4. right**: mescla todas as linhas de x que estiverem contidas em y.<br>

#### 1. inner

![](https://s3-sa-east-1.amazonaws.com/lcpi/00764540-d97a-496a-96e7-e0418da993a4.png)

```python
df_ineer = df_customer.merge(df_info, on="id", how="inner")
```

```python
df_ineer
```

#### 2. outer

![](https://s3-sa-east-1.amazonaws.com/lcpi/25e34804-6980-41a4-a95b-1628c6ecc6f6.png)

```python
df_outer = df_customer.merge(df_info, on="id", how="outer")
df_outer
```

#### 3. left

![](https://s3-sa-east-1.amazonaws.com/lcpi/d6fd1e43-7e7c-405c-983c-434e224bb868.png)

```python
df_left = df_customer.merge(df_info, on="id", how="left")
df_left
```

#### 4. right

![](https://s3-sa-east-1.amazonaws.com/lcpi/d3349192-e04a-4d9b-b3a0-8ff9c82f8bd4.png)

```python
df_right = df_customer.merge(df_info, on="id", how="right")
df_right
```

1. Qual o percentual de graduados empregados?

2. Qual o percentual de graduados e empregados que conseguiram empréstimo?

4. Entre homens e mulheres quem conseguiu mais empréstimos?

5. Qual o percentual de dados faltantes?

6. Quais outras variáveis podem ser criadas com base no dataset?

7. Qual o perfil de aprovados?

```python
import pandas as pd
df = pd.read_csv("clientes.csv")
```

```python
df.head()
```

```python
# Questão 1. Qual o percentual de graduados empregados?

filtro1 = df["empregado"] == "Yes"
filtro2 = df["educacao"] == "Graduate"

num_grd_emp = len(df[filtro1 & filtro2])
percentual = num_grd_emp/len(df)
percentual = percentual*100
print(f"{round(percentual, 2)} %")
```

```python
df[["empregado", "educacao"]].groupby("educacao").count()
```

```python
# Questão 2. Qual o percentual de graduados e empregados que conseguiram empréstimo?

filtro1 = df["empregado"] == "Yes"
filtro2 = df["educacao"] == "Graduate"
filtro3 = df["aprovacao_emprestimo"] == "Y"

num_grd_emp = len(df[filtro1 & filtro2 & filtro3])
percentual = num_grd_emp/len(df)
percentual = percentual*100
print(f"{round(percentual, 2)} %")
```

```python
# 3. Entre solteiros e casados, quem conseguiu mais empréstimos?

# Solteiros
filtro1 = df["estado_civil"] == "No"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Solteiros: {round(percentual, 2)} %")

# Casados
filtro1 = df["estado_civil"] == "Yes"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Casados: {round(percentual, 2)} %")
```

```python
# 4. Entre homens e mulheres quem conseguiu mais empréstimos?

# Homens
filtro1 = df["sexo"] == "Male"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Homens: {round(percentual, 2)} %")

# Mulheres
filtro1 = df["sexo"] == "Female"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Mulheres: {round(percentual, 2)} %")
```

```python
# Resolvendo de forma proporcional:

print("Proporção de homens e mulheres no df:")
print(df["sexo"].value_counts(normalize=True))

print("\nProporção de homens e mulheres aprovados:")
print(df[df["aprovacao_emprestimo"] == "Y"]["sexo"].value_counts(normalize=True))

# A proporção para ambos os sexos em ambos os casos é praticamente a mesma.
```

```python
# 5. Qual o percentual de dados faltantes?

cel_nulas = df.isnull().sum().values.sum()
cel_total = df.columns.values.shape[0] * len(df)
print(f"Percentual de dados faltantes: {round((cel_nulas/cel_total)*100, 2)} %")
```

```python
# 6. Quais outras variáveis podem ser criadas com base no dataset?

# Valor total de emprestimo
df["total_emprestimo"] = df["emprestimo"] * df["prestacao_mensal"]
```

```python
# Categorizar faixas de renda

s_min =  1320

categorias = []
for r in df["renda"].values:
  if r < 3*s_min:
    categorias.append("E")
  elif r < 6*s_min:
    categorias.append("D")
  elif r < 9*s_min:
    categorias.append("C")
  elif r < 12*s_min:
    categorias.append("B")
  else:
    categorias.append("A")

df["faixa_renda"] = categorias
```

```python
# 7. Qual o perfil de aprovados?

aprovados = df[df["aprovacao_emprestimo"] == "Y"]

# Relacionar aprovados com faixa de renda
# Relacionar aprovados com sexo
# Relacionar aprovados com estado_civil
# Relacionar aprovados com educacao
# Relacionar aprovados com empregado
# Numeros de imoveis por tipo
# Relacionar aprovação com histórico
```

```python
# Aprovados por faixa de renda:

aprovados["faixa_renda"].value_counts()
```

```python
features = ['sexo', 'estado_civil', 'dependentes', 'educacao', 'empregado', 'historico_credito', 'imovel','faixa_renda']

for caracteristica in features:
  print(f"Feature: {caracteristica}:")
  print(aprovados[caracteristica].value_counts(normalize=True))
  print()
```
